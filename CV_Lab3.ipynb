```python

# import Opencv
import cv2
import matplotlib.pyplot as plt
  
# import Numpy
import numpy as np
  
# read a image using imread
img = cv2.imread('/content/drive/MyDrive/CV/Images/pexels-photo-1632028.jpeg', 0)
  


histogram, bin_edges = np.histogram(img, bins=256, range=(0, 1))
print("HIstogram is :")
print(histogram)
print("Edges are:")
# creating a Histograms Equalization
# of a image using cv2.equalizeHist()
equ = cv2.equalizeHist(img)
histogram, bin_edges = np.histogram(equ, bins=256, range=(0, 1))
print("HIstogram is :")
print(histogram)
# stacking images side-by-side
res = np.hstack((img, equ))
  
# show image input vs output
plt.imshow(img)
plt.show()
plt.imshow(equ)
plt.show()
plt.imshow(res)
plt.show()
```
