```python
# Open image using openCV2
opencv_image = cv.imread("/content/indianteam.png")
  
# Notice the COLOR_BGR2RGB which means that the color is
# converted from BGR to RGB
color_coverted = cv.cvtColor(opencv_image, cv.COLORMAP_HOT)
plt.imshow(color_coverted)
plt.show()
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)
plt.imshow(color_coverted)
plt.show()
```

```python
# Rotate Image By 180 Degree
rotated_image1 = image1.rotate(180)
 
# This is Alternative Syntax To Rotate
# The Image
rotated_image2 = image1.transpose(Image.ROTATE_90)
 
# This Will Rotate Image By 60 Degree
rotated_image3 = image1.rotate(60)
 
# display(rotated_image1)
# display(rotated_image2)
# display(rotated_image3)

display(rotated_image1,rotated_image2,rotated_image3)
```

```python
# Size of the image in pixels (size of original image)
# (This is not mandatory)
width, height = image1.size
 
# Setting the points for cropped image
left = 5
top = height / 4
right = 164
bottom = 3 * height / 4
 
# Cropped image of above dimension
# (It will not change original image)
im1 = image1.crop((left, top, right, bottom))
 
# Shows the image in image viewer
im1
```
